```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
training = pd.read_csv('train_data.csv',header=None)
label = training[17].values
training = training.drop(17,axis=1)
training = training.drop(0,axis=1)

testing = pd.read_csv('test_data.csv',header=None)
test_idx = testing[0].values
testing = testing.drop(0,axis=1)
training_len = len(training)
display(testing)
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder


label_le = LabelEncoder()
label = label_le.fit_transform(label)

all_data = pd.concat([training,testing])

catego_features = [2,3,4,5,7,8,9,11,16]

catego_le = LabelEncoder()

num_values = []
for i in catego_features:
    all_data[i] = catego_le.fit_transform(all_data[i].values)
    classes_list = catego_le.classes_.tolist()
    
    # store the total number of values
    num_values.append(len(classes_list))
    
    # replace '?' with 'NaN'
    if 'unknown' in classes_list:
        idx = classes_list.index('unknown')
        all_data[i] = all_data[i].replace(idx, np.nan)

display(all_data.head(15))
```

```python
import impyute as impy

imputed_data = impy.imputation.cs.mice(all_data.values)
#imr = imr.fit(all_data.values)
#imputed_data = imr.transform(all_data.values)
```

```python
training_data = imputed_data[:training_len]
testing_data =imputed_data[training_len:]
# label
```

```python
from sklearn.preprocessing import StandardScaler
'''
Standardlize numeric feature
'''
std = StandardScaler()
std.fit(training_data)
training_data = std.transform(training_data)
testing_data = std.transform(testing_data)
```

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state = 0)
training_data, label = rus.fit_sample(training_data, label)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

rf = RandomForestClassifier(n_estimators=1000 , n_jobs=4 , random_state=386)
rf.fit(training_data , label)
prediction = rf.predict(testing_data)
```

```python
with open('sample_output.csv','w') as f:
    f.write('Id,Prediction\n')
    for idex , p in zip(test_idx,prediction):
        f.write('{},{}\n'.format(idex,p))
```
